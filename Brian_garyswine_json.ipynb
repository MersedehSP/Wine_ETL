```python
import requests
import json
import pandas as pd
import time
```

```python
# api call 
# should take about 5 mins
unit5 = str(0)
real_data5=[]


for x in range(1,134):
    #check running count
    print(unit5)
    #sleep just in case the api has a limitation on amount of calls
    time.sleep(2)

    url =  f"https://garyswine.com/api/v1/products/search.json?views%5B%5D=Wine&include_price_range=true\
    &merchant_id=5963aaf8155cb61a06cfe5de&limit=18&skip={unit5}&api_key=e0d3a091dc0d81547d6e168be2b3492a&sdk_guid=\
    ee1de86b-1cba-1ffd-1117-a590a7bd0071&client_origin=app%3A%2F%2Fstorefront.5963aaf8155cb61a06cfe5de&new_style=true"


    
    response5 = requests.get(url)
    data5 = response5.json()
    #print(url)
    #gets the main chunk of data
    product5 = data5["data"]["products"]
    #store the pages infomration into a list
    names5 = []
    #moves the api call to the next page of bottles
    unit5 = x*18
    
    for name in product5:
        names5.append(name)
    
    
    #in each bottle get information    
    for x in names5:
        name5= x["name"]
        try:
            year5 = x["additional_properties"]["age"]
        except KeyError:
            year5 = "NA"
        try:
            country5 = x["additional_properties"]["country"]
        except KeyError:
            country5 ="NA"    
        try:
            region5 = x["additional_properties"]["region"]
        except KeyError:
            region5 ="NA" 
        try:
            sku5 = x["additional_properties"]["sku"]
        except KeyError:
            sku5 ="NA" 
        try:
            state5 = x["additional_properties"]["state"]
        except KeyError:
            state5 ="NA" 
        try:
            subtype5 = x["additional_properties"]["subtype"]
        except KeyError:
            subtype5 ="NA" 
        try:
            type5 = x["additional_properties"]["type"]
        except KeyError:
            type5 ="NA" 
        try:
            varieta5 = x["additional_properties"]["varietal"]
        except KeyError:
            varieta5 ="NA"
        try:
            description5 = x["description"]
        except KeyError:
            description5 ="NA"
        try:
            merchant5 = x["merchants"][0]["merchant_name"]
        except KeyError:
            merchant5 ="NA"
        try:
            address5 = x["merchants"][0]["full_address"]
        except KeyError:
            address5 ="NA"


        price5 = x["merchants"][0]["product_options"][0]["price"]

        # add to main list the below information
        real_data5.append({"name" : name5,
                       "age" : year5,
                       "country" : country5,
                       "region" : region5,
                        "sku" : sku5,
                        "state" : state5,
                        "subtype" : subtype5,
                        "type" : type5,
                        "varietal" : varieta5,
                        "price" : price5,
                        "description" : description5,  
                        "merchant" : merchant5,
                        "address" : address5
                       })

    #print(test123)

#print(json.dumps(real_data5, indent=4, sort_keys=True))    
   
    

print(len(real_data5))
```
