```python

import time
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from splinter import Browser
```

```python
# Import Splinter and set the chromedriver path

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://www.petrocksliquors.com/category_Wine"
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
petrocks_url = "https://www.petrocksliquors.com/category_Wine"
browser.visit(petrocks_url)

wine_list = []

for i in range (5):    
     
    
            for j in range(20):
                time.sleep(2)
                wine ={}
            
                browser.find_by_tag('div[style="width:200px; height:200px;overflow: hidden;"]')[j].click()
                wine_html =  browser.html
                wine_soup = BeautifulSoup(wine_html, 'html.parser')
                

                try:

                    try:
                        name = wine_soup.find("span", id="bmg_itemdetail_prod_title").text
                    except:
                        name:"name"  
                        browser.back()    
                    try: 
                        country =wine_soup.find_all("a", class_="reglink")[0].text
                    except:
                        country = "N/A"
                    try:    
                        region = wine_soup.find(text= "Region:").findNext("a").contents[0]
                    except:
                        region = "N/A"
                    try:
                        sub_region = wine_soup.find(text= "Sub-Region:").findNext("a").contents[0]
                    except:

                        sub_region = "N/A"
                    try:

                        grape_varietal = wine_soup.find(text= "Grape Varietal:").findNext("a").contents[0]
                    except:
                        grape_varietal = "N/A"
                    try:    
                        wine_type = wine_soup.find(text= "Type:").findNext("a").contents[0]
                    except:
                        wine_type = "N/A"
                    try:    
                        bottle_price = wine_soup.find("span", class_="RegularPrice").text
                    except:
                        bottle_price = "N/A"
                    try:
                        sale_price = wine_soup.find("span", class_="SalePrice").text
                    except:
                        sale_price = "N/A"
                    try:    
                        description = wine_soup.find("div", id="moreinfo")
                        description = description.find_all("tr")[1].text
                    except:
                        description = "N/A"
                    merchant = "Petrocks Liquors"
                    address = "419 Amwell Road Hillsborough, NJ 08844"
                except:
                    name = None
                    country =None
                    region = None
                    sub_region = None
                    grape_varietal = None
                    wine_type = None
                    bottle_price = None
                    sale_price = None
                    description = None
                    merchant = None
                    address = None
                    
                wine={
                     "name":name, 
                     "country": country,
                     "region": region,
                    "sub_region": sub_region,
                     "grape_varietal" : grape_varietal,
                      "wine_type" : wine_type,
                       "bottle_price":bottle_price,
                       "sale_price":sale_price,
                       "description" : description, 
                       "merchant": merchant,
                       "address" :address
                     }

                wine_list.append(wine)
                print(name)    
                browser.back()
            browser.find_by_text(">>").click()
            print(f"I clicked on next page :D", {i})
        
wine_list
```

```python
file_path = "petrocks_finalllll4k.csv"
wine_df = pd.read_csv(file_path, encoding='ISO-8859-1')
wine_df.head(80)
```

```python
#renaming the columns to matc with everyone
wine_df.rename(columns = {'region':'state'}, inplace = True) 
wine_df.rename(columns = {'sub_region':'region'}, inplace = True)
wine_df.rename(columns = {'grape_varietal':'varietal'}, inplace = True) 
wine_df.rename(columns = {'sale_price':'price'}, inplace = True) 
```

```python
#dropping unwanted columns
#wine_df.drop(columns = {'region':'state'}, inplace = True) 
wine_df = wine_df.drop(['wine_type','bottle_price',], axis='columns')
```

```python
#droppinig the empty rows
wine_df = wine_df.dropna(how = "all")
wine_df
```
