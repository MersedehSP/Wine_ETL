```python
import pandas as pd
```

```python
# Save path to data set in a variable
data_file = "garys_wine_real2.csv"
```

```python
# Use Pandas to read data:
wine_df = pd.read_csv(data_file)
```

```python
#gives you the top 5 rows of the file
wine_df.head()
```

```python
#From below you can play with the list to reorganize the columns(Wine Table)
organized_wine_df = wine_df[["name", "price", "description", "varietal", "region"]]
organized_wine_df.head()
```

```python
#From below you can play with the list to reorganize the columns(Country)
country_wine_df = wine_df[["country", "region", "state", "address"]]
country_wine_df.head()
```

```python
#From below you can play with the list to reorganize the columns(Merchant)
#Merchant ID
merchant_wine_df = wine_df[["merchant", "address"]]
merchant_wine_df.head()
```

```python
#From below you can play with the list to reorganize the columns(Merchant_wine)
merchant_wine_wine_df = wine_df[["name", "merchant"]]
merchant_wine_wine_df.head()
```

```python
#From below you can play with the list to reorganize the columns(Style)
style_wine_df = wine_df[["varietal"]]
style_wine_df.head()
```

```python
#Move a NEW DataFrame into a CSV file:
style_wine_df.to_csv("style_wine.csv", index=False, header=True)
merchant_wine_wine_df.to_csv("merchant_wine_wine.csv", index=False, header=True)
style_wine_df.to_csv("style_wine.csv", index=False, header=True)
style_wine_df.to_csv("style_wine.csv", index=False, header=True)
```
