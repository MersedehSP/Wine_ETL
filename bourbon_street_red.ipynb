```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = "https://www.bourbonstreetwineandspirits.com/search/categories/Red%20Wine/result_size/96"
browser.visit(url)
```

```python
wine_list = []
wine = {}

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

links=soup.find_all('a', class_='product-link')
for link in links:
    urls= 'https://www.bourbonstreetwineandspirits.com'+(link.get('href'))
    #print(urls)
    #for detail_url in urls:
     #   print (detail_url)
    detail_html=urlopen(urls)
    soup_detail = BeautifulSoup(detail_html, 'lxml')
    #print(soup_detail)
    details= soup_detail.find_all(class_={'product-name mb-0','btn btn-outline-primary btn-block update-quantity-input', 'row bg-light py-3 mb-3'})
    #print(details)
             
    name = soup_detail.find('h1', class_="product-name").text        
    #print(name)

    try:
        price = soup_detail.find('button', class_="btn btn-outline-primary btn-block update-quantity-input")["data-price"]
    except TypeError:
        price = "N/A"
    #print(price)
    
    region = soup_detail.find("table").find_all('a')[3].text
    #print(region)
    
    wine_type = soup_detail.find("table").find_all('a')[0].text
    #print(wine_type)
    
    varietal = soup_detail.find("table").find_all('a')[1].text
    #print(varietal)
    
    country = soup_detail.find("table").find_all('a')[2].text
    #print(country)
    
    merchant = "bourbon Street Wine and Spirits"
    address = "1217 Route 22 West, Phillipsburg NJ 08865"
    
    wine={
        "name":name,
        "price":price,
        "region":region,
        "wine_type":wine_type,
        "varietal":varietal,
        "country":country,
        "merchant":merchant,
        "address":address
    }
    
    wine_list.append(wine)

wine_list
  
                    
    
    
        




    
    
   
```
