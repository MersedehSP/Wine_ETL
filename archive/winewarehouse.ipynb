```python
from bs4 import BeautifulSoup
from splinter import Browser
import requests
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {"executable_path": ChromeDriverManager().install()}
browser = Browser("chrome", **executable_path, headless=False)

url= "https://winewarehousenj.com/shop/?container-id=5ce3fe41480ec357ee9b9d7a&title=Shop%2520Wine"

browser.visit(url)
    
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
soup
```

```python
grid = browser.find_by_tag('div[class="ch-product-name"]')

for x in grid:
    print(x)


print(grid[0])

#<div _ngcontent-kjl-c53 class="ch-product-name"> 19 Crimes Cabernet Sauvignon </div>
```
