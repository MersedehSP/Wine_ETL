```python
from bs4 import BeautifulSoup
from splinter import Browser
import requests
import pandas as pd
import time
from webdriver_manager.chrome import ChromeDriverManager
```

```python
executable_path = {"executable_path": ChromeDriverManager().install()}
browser = Browser("chrome", **executable_path, headless=False)

url= "https://www.totalwine.com/wine/c/c0020?viewall=true&pageSize=24&department=Wine"

browser.visit(url)
    
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
soup.prettify()
```

```python
#browser.find_by_tag('div[class="titleRatingsHolder__1TZlqhxF"]')
#anclick="view_as_list"
#browser.find_by_text('Switch to List View').click()
listbutton = browser.find_by_tag('button[anclick="view_as_list"]')
listbutton.click()
```

```python
#product_list= soup.find_all('article', class_= 'productCard__2nWxIKmi')
product_list= browser.find_by_tag('article[class="productCard__2nWxIKmi"]')


#print(product_list)
bottles=[]
for bottle in product_list:
    h2= bottle.find_by_tag('h2')
    div_text_1 = bottle.find_by_tag('div')[2]
    div_text_2 = div_text_1.find_by_tag('div')[4]
    a1=div_text_2.find_by_tag('a')[1].text
    #name = h2.text
    bottles.append({"name" : h2.text,
                    "region" : a1} )
    break
    #print(h2.text)
print(bottles)

```
