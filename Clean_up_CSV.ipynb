```python
import pandas as pd
```

```python
# Save path to data set in a variable
data_file_1 = "garys_wine_real2.csv"
data_file_2 = "chathambuyritecleaned.csv"
data_file_3 = "petrocks_dummy.csv"
```

```python
# Use Pandas to read data:
garys_df = pd.read_csv(data_file_1)
bottleking_df = pd.read_csv(data_file_2)
petrocks_df = pd.read_csv(data_file_3)
```

```python
#gives you the top 5 rows of the file
garys_df.head()
```

```python
#gives you the top 5 rows of the file
bottleking_df.head()
```

```python
#From below you can play with the list to reorganize the columns(Wine Table)
bottleking_df = bottleking_df[["name", "country", "region", "state", "varietal", "price", "description","merchant","address"]]
bottleking_df.head()
```

```python
#Find all instances of a value and replace
bottleking_df["price"] = bottleking_df["price"].str.replace("$", "").astype(float)
```

```python
#gives you the top 5 rows of the file
petrocks_df.head()
```

```python
petrocks_df = petrocks_df[["name", "country", "region", "state", "varietal", "price", "description","merchant","address"]]
petrocks_df.head()
petrocks_df["price"] = petrocks_df["price"].str.replace("$", "").astype(float)
```

```python
petrocks_df.head()
```

```python
wineall_df=garys_df.append(bottleking_df,ignore_index=True)
wineall_df=wineall_df.append(petrocks_df,ignore_index=True)
wineall_df
```
